```python
import pandas as pd
df = pd.read_csv('Chicago_Illinois_US.csv')
df.info()
```

```python
filtroA = df[(df["host_name"] == "Rebecca") | (df["host_name"] == "Joe") | (df["host_name"] == "Jillian") | (df["host_name"] == "Meighan") | (df["host_name"] == "Jb")]
filtroA
```

```python
#Convertir filtro a csv
filtroA.to_csv("FiltroA_FiltroPorObjeto_Conector_o_5_host_name.csv")
```

```python
#La columna a filtrar no es de tipo date por lo que primero la convertiremos
df["host_since"] = pd.to_datetime(df["host_since"], errors="coerce")
df.info()
```

```python
filtroB = df[df["host_since"] < "01/01/2019"]
filtroB
```

```python
#Convertir filtro a csv
filtroB.to_csv("FiltroB_FiltroPorComparacion_host_since_menor_01_01_2019.csv")
```

```python
filtroC = df[df["host_response_time"] == "within a day"]
filtroC
```

```python
#Convertir filtro a csv
filtroC.to_csv("FiltroC_FiltroPorObjeto_host_response_time_within a day.csv")
```

```python
filtroD = df[df["accommodates"] >= 5]
filtroD
```

```python
#Convertir filtro a csv
filtroD.to_csv("FiltroD_FiltroPorComparacion_accommodates_MayorIgual_5.csv")
```

```python
filtroE = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtroE
```

```python
#Convertir filtro a csv
filtroE.to_csv("FiltroE_FiltroPorObjeto_conector_o_room_type_Entire home-apt_o_Private room_.csv")
```
