```python
#Importamos la librerias requeridas
import pandas as pd
```

```python
#Carga de un archivo .xlsx sin indice 
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto
filtro1 = df[df["NOMBRE_CLIENTE"] == "REVUELTA MONTECARLO"]
filtro1
```

```python
#Converti archivo filtrado a CSV
filtro1.to_csv("Entregable1_FiltroPorObjeto_NOMBRE_CLIENTE.csv")
```

```python
#Filtro por filas
#filtro2 = df.iloc[55:61, : ]       #Filas consecutivas
filtro2 = df.iloc[[70,80,90], : ]         #Filas no consecutivas
filtro2
```

```python
#Converti archivo filtrado a CSV
filtro2.to_csv("Entregable2_FiltroPorFilas_70_80_90.csv")
```

```python
#Filtro por columnas
#filtro3 = df.iloc[ : , 10:13]           #Columnas consecutivas
filtro3 = df.iloc[ : , [2,3,13]]   #Columnas no consecutivas
filtro3
```

```python
#Converti archivo filtrado a CSV
filtro3.to_csv("Entregable3_FiltroPorColumnas_2_3_13.csv")
```

```python
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=0)
df1
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro7 = df[df["PRECIO_UNITARIO"] <= 800]
filtro7
```

```python
#Converti archivo filtrado a CSV
filtro7.to_csv("Entregable7_FiltroPorComparacion_PRECIO_UNITARIO_MenorIgual_800.csv")
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#Filtro Y
filtro8 = df[(df["PRECIO_UNITARIO"] < 2000) & (df["MARGEN_TOTAL_CALCULADO"] > 3000)]
filtro8
```

```python
#Converti archivo filtrado a CSV
filtro8.to_csv("Entregable8_FiltroPorComparacion_Conector_Y_PRECIO_UNITARIO_Menor_2000_Y_MARGEN_TOTAL_CALCULADO_Mayor_3000.csv")
```

```python
#Filtro o
filtro9 = df[(df["NOMBRE_VENDEDOR"] == "AARON NUÑO CORDOVA") | (df["NOMBRE_CLIENTE"] == "LUIS ENRIQUE PATIÑO PEREZ")]
filtro9
```

```python
#Converti archivo filtrado a CSV
filtro9.to_csv("Entregable9_FiltroPorComparacion_Conector_O_NOMBRE_VENDEDOR_AARON NUÑO CORDOVA_O_NOMBRE_CLIENTE_LUIS ENRIQUE PATIÑO PEREZ.csv")
```

```python
#Filtro not
filtro10 = df[~(df["CANT"] > 5)]
filtro10
```

```python
#Converti archivo filtrado a CSV
filtro10.to_csv("Entregable10_FiltroPorComparacion_Conector_Negacion_CANT_Mayor_5.csv")
```
